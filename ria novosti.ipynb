```python
import pandas as pd
import requests
from datetime import datetime
import re
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import time
import numpy as np
from iteration_utilities import flatten
```

```python
dates=pd.date_range('2013-01-01',periods=3388,freq='D')
dates=list(map(lambda x: str(x).split(' ')[0],dates))
dates=list(map(lambda x: x.split('-'),dates))
dates=list(map(lambda x: "".join(x),dates))
hrefs=[]
```

```python
news=[]
dates_news=[]
headers=[]
```

```python
start=time.time()
for i in dates:
    time.sleep(np.random.uniform(0,1.5))
    url="https://ria.ru/services/economy/more.html?id=1784460664&date="+i+"T122939"
    response=requests.get(url)
    soup=BeautifulSoup(response.text)
    links=soup.find_all('div',class_="list-item__content")
    for j in links:
        link=j.find_all('a',href=True)[0]['href']
        hrefs.append(link)
        print(time.time()-start)
        print(i)
```

```python

for href in hrefs:
    time.sleep(np.random.uniform(0,1.5))
    url=href
    response=requests.get(url)
    soup=BeautifulSoup(response.text)
    texts=soup.find_all('div',class_='article__text')
    try:
        texts=texts[0].find_all('p',class_='doc__text')
        clean_text=[]
        for text in texts:
            clean_text.append(re.sub("<.*?>","",str(text)))
        news.append(' '.join(clean_text)) 
        date=url.split('/')[-2]
        dates_news.append(date)
        header=soup.find_all('h1',class_='article__title')
        header=re.sub("  ","",re.sub('\r.*?\n','',re.sub("<.*?>","",str(header[0]))))
        headers.append(header)
    except IndexError:
            continue
```

```python

for href in hrefs:
    time.sleep(np.random.uniform(0,1.7))
    url=href
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')

    try:
        texts=soup.find_all('p')
        l=[x.find_all('p') for x in texts]
        m=list(flatten(l))
        text=" ".join([str(x) for x in m])
        news.append(re.sub("<.*?>","",text))
        date=url.split('/')[-2]
        dates_news.append(date)
        header=soup.find_all('h1',class_='article__title')
        header=re.sub("  ","",re.sub('\r.*?\n','',re.sub("<.*?>","",str(header[0]))))
        headers.append(header)
    except IndexError:
        continue
```
